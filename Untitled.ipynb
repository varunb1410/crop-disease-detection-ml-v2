```python
import tensorflow as tf
import json

# Load trained model
model = tf.keras.models.load_model("plant_disease_model_final.keras")

# Load class names
with open("class_names.json") as f:
    class_names = json.load(f)

print("Model loaded successfully")
```

```python
import os
from PIL import Image
import numpy as np

folder = r"D:\major project\crop_disease_prediction_v2\dataset\valid\Tomato___Late_blight"

# get first image in folder
file_name = os.listdir(folder)[0]

img_path = os.path.join(folder, file_name)

print("Using image:", img_path)

img = Image.open(img_path).convert("RGB")
img = img.resize((128,128))

img_array = np.array(img)
img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)

predicted_class = class_names[np.argmax(prediction)]
confidence = np.max(prediction)

print("Prediction:", predicted_class)
print("Confidence:", confidence)
```
